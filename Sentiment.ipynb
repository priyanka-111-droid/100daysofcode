```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
pip install datasets
```

```python

# Load the dataset
from datasets import load_dataset
dataset = load_dataset("carblacac/twitter-sentiment-analysis")
# Access the training data
train_data = dataset["train"]
# Access the testing data
test_data = dataset["test"]

# Access the 'text' and 'feeling' columns in the training set
texts_train = train_data['text']
labels_train = train_data['feeling']

# Access the 'text' and 'feeling' columns in the testing set
texts_test = test_data['text']
labels_test = test_data['feeling']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(texts_train, labels_train, test_size=0.2, random_state=42)
```

```python
# Create a Naive Bayes classifier pipeline
model_nb = make_pipeline(CountVectorizer(), MultinomialNB())
```

```python
# Train the model
model_nb.fit(X_train, y_train)
```

```python
# Make predictions on the test set
predictions_nb = model_nb.predict(X_test)
```

```python
# Evaluate the model
accuracy_nb = accuracy_score(y_test, predictions_nb)
print(f"Naive Bayes Accuracy: {accuracy_nb}")
print(classification_report(y_test, predictions_nb))
```
